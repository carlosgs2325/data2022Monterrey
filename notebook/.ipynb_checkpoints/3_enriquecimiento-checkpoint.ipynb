```python
%matplotlib inline
import pandas as pd
```

```python
compras_df=pd.read_csv('data/clean_compras.csv')
```

```python
compras_df.head()
```

```python
compras_df.dtypes
```

```python
compras_df['FECHA'].unique
```

```python
compras_df['FECHA']=pd.to_datetime(compras_df['FECHA'])
```

```python
compras_df['DIA']=compras_df['FECHA'].apply(lambda x: x.day)
```

```python
compras_df['MES']=compras_df['FECHA'].apply(lambda x: x.month)
```

```python
compras_df['AÑO']=compras_df['FECHA'].apply(lambda x: x.year)
```

# Proyectar una variable con otra

```python
sheet_sales=compras_df['SHEET'].value_counts().to_dict()
```

```python
list(sheet_sales.keys())[0:5]
```

```python
compras_df['SHEET_SALES']=compras_df['SHEET'].apply(lambda x: sheet_sales[x])
```

```python
compras_df.head()
```

```python
ben_mean_compras=compras_df.groupby('BENEFICIARIO')['IMPORTE'].mean()
```

```python
ben_mean_compras
```

```python
ben_mean_compras.to_dict()
```

```python
compras_df['IMPORTE_PROMEDIO_BEN']=compras_df['BENEFICIARIO'].apply(lambda x: ben_mean_compras[x])
```

```python
compras_df.to_csv('data/compras_df_enriched.csv', index=None)
```
